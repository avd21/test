#Imports

```python
import seaborn as sns
```

```python
tips= sns.load_dataset('tips')
```

#DATA
Seaborn comes with built in data sets

```python
tips.head()
```

#distplot
displot() shows the distribution of a univariate set of observation.

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=50)
```

#**jointplot**
jointplot() allowss you to basically much up two distplots for bivariate data.With your choice of what **kind** parameter to campare with:


*   scatter
*   hex

*   reg
*   resid

*   kde







```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

#**pairplot**
pairplot() will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex",palette='coolwarm')
```

#**rugplot**
rugplots() are actually a very simple concept they just draw a dash mark for every point on a univariate distribution.They are the building blocks of the KDE plot.

```python
sns.rugplot(tips['total_bill'])
```

#**CATEGORIAL** **PLOTS**

#**barplot**
barplot() is a general plot that allows you to aggregate the categorial data based of some funcion by default the mean.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

#**countplot**
countplot() is same as barplot except the estimator is explicitly counting the no. of occurrences.Which is why we only pass the x value.

```python
sns.countplot(x='sex',data=tips)
```

#**boxplot**
boxplot() shows the distribution of quantitative data in a way that facilitates comparisons between variables or accross level of a categorial variable.

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(data=tips,orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow',hue='smoker')
```

#**violinplot**
violinplot() shows the distribution of quantitative data across several levels of one categorical variables such that those distribution can be compared.

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='Set1',hue='sex')
```

#stripplot

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,palette='Set1',hue='sex')
```

#**swarmplot**

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,palette='Set1',hue='sex')
```

#**factorplot**
factorplot() is the most general form of a categorial plot.It can take in a **kind** parameter to adjust the plot type

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
